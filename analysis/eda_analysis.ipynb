## Step 3: Chip Manufacturing

### Q1: Chip manufacture (Dataset 4)

#### 1. Heatmap: Market Share of Semiconductor Inputs by Country (2024)

```python
provision_df = pd.read_csv('data/processed/dataset4/provision_clean.csv')
providers_df = pd.read_csv('data/processed/dataset4/providers_clean.csv')

merged_df = pd.merge(provision_df, providers_df[['provider_name', 'country']], on = 'provider_name', how = 'left')
merged_df['country'] = merged_df.apply(
    lambda row: row['provider_name'] if pd.isna(row['country']) else row['country'], axis = 1
)
merged_2024_df = merged_df[merged_df['year'] == 2024].drop_duplicates(
    subset = ['provider_name', 'provided_name', 'share_provided']
)

top15_inputs = merged_2024_df.groupby('provided_name')['share_provided'].sum().nlargest(15).index.tolist()
top8_countries = merged_2024_df.groupby('country')['share_provided'].sum().nlargest(8).index.tolist()

heatmap_df = merged_2024_df[
    (merged_2024_df['provided_name'].isin(top15_inputs)) &
    (merged_2024_df['country'].isin(top8_countries))
].groupby(['provided_name', 'country'])['share_provided'].sum().reset_index()

heatmap = alt.Chart(heatmap_df).mark_rect().encode(
    alt.X('country:N', title = "Country", axis = alt.Axis(labelAngle = 0)),
    alt.Y('provided_name:N', title = "Input"),
    alt.Color('share_provided:Q',
              scale = alt.Scale(scheme = 'blues'),
              title = "Market Share (%)"),
    tooltip = [
        alt.Tooltip('provided_name:N', title = "Input"),
        alt.Tooltip('country:N', title = "Country"),
        alt.Tooltip('share_provided:Q', title = "Market Share (%)", format = '.1f')
    ]
).properties(
    width = 450,
    height = 450,
    title = "Market Share of Semiconductor Inputs by Country (2024)"
)

heatmap
```

#### 2. Heatmap: Average Market Share of Manufacturing Stage by Country (2024)

```python
provision_df = pd.read_csv('data/processed/dataset4/provision_clean.csv')
providers_df = pd.read_csv('data/processed/dataset4/providers_clean.csv')
sequence_df = pd.read_csv('data/processed/dataset4/sequence_clean.csv')

merged_df = pd.merge(provision_df, providers_df[['provider_name', 'country']], on = 'provider_name', how = 'left')
merged_df['country'] = merged_df.apply(
    lambda row: row['provider_name'] if pd.isna(row['country']) else row['country'], axis = 1
)
merged_2024_df = merged_df[merged_df['year'] == 2024]

joined_df = pd.merge(
    merged_2024_df,
    sequence_df[['input_name', 'goes_into_name']],
    left_on = 'provided_name',
    right_on = 'input_name',
    how = 'inner'
).dropna(subset = ['goes_into_name'])

top_countries = joined_df.groupby('country')['share_provided'].sum().nlargest(8).index.tolist()
joined_filtered_df = joined_df[joined_df['country'].isin(top_countries)]

heatmap_df = joined_filtered_df.groupby(
    ['goes_into_name', 'country']
)['share_provided'].mean().reset_index()

heatmap = alt.Chart(heatmap_df).mark_rect().encode(
    alt.X('country:N', title = "Country", axis = alt.Axis(labelAngle = 0)),
    alt.Y('goes_into_name:N', title = "Manufacturing Stage"),
    alt.Color('share_provided:Q',
              scale = alt.Scale(scheme = 'blues'),
              title = "Avg Market Share (%)"),
    tooltip = [
        alt.Tooltip('goes_into_name:N', title = "Stage"),
        alt.Tooltip('country:N', title = "Country"),
        alt.Tooltip('share_provided:Q', title = "Avg Market Share (%)", format = '.1f')
    ]
).properties(
    width = 450,
    height = 450,
    title = "Average Market Share of Manufacturing Stage by Country (2024)"
)

heatmap
```

### Q2: Chip price across AI milestones (Dataset 5)

```python
chip_price_df = pd.read_csv('data/processed/dataset5/chip_price_clean.csv')
chip_price_df['observation_date'] = pd.to_datetime(chip_price_df['observation_date'])

events_df = pd.DataFrame({
    'date': ['2018-01-01', '2022-11-01'],
    'label': ['OpenAI Founded (2018)', 'ChatGPT Released (Nov 2022)']
})

line = alt.Chart(chip_price_df).mark_line().encode(
    alt.X('observation_date:T', title = "Date"),
    alt.Y('Price_Index:Q', title = "Price Index", scale = alt.Scale(zero = False)),
    tooltip = [
        alt.Tooltip('observation_date:T', title = "Date"),
        alt.Tooltip('Price_Index:Q', title = "Price Index", format = '.2f')
    ]
)

rules = alt.Chart(events_df).mark_rule(
    strokeDash = [6, 4]
).encode(
    alt.X('date:T'),
    alt.Color('label:N', title = "Event", scale = alt.Scale(range = ['red', 'blue']))
)

labels = alt.Chart(events_df).mark_text(
    align = 'left',
    dx = 5,
    dy = -130,
    fontSize = 11
).encode(
    alt.X('date:T'),
    alt.Color('label:N', legend = None, scale = alt.Scale(range = ['red', 'blue'])),
    text = 'label:N'
)

line_chart = (line + rules + labels).properties(
    width = 600,
    height = 350,
    title = "Chip Price Index Before and After Key AI Milestones (2015-2025)"
)

line_chart
```

### Q1: Money funds (Dataset 6, 7)

#### 1. Bar Chart: Total Semiconductor Investment by Country and Activity (2018-2022)

```python
oecd_df = pd.read_csv('data/processed/dataset6/oecd_investment_clean.csv')

fx_rates = {
    'Korea': 1 / 1200,
    'Japan': 1 / 110,
    'Chinese Taipei': 1 / 30,
    'China (People\u2019s Republic of)': 1 / 6.7,
    'United States': 1,
    'Germany': 1.1,
    'France': 1.1,
    'United Kingdom': 1.3
}

oecd_df['investment_usd'] = oecd_df.apply(
    lambda row: row['Investment amount'] * fx_rates.get(row['Country'], 1), axis = 1
)

activity_labels = {
    'Manufacture of electronic components and boards': 'Electronic Components',
    'Manufacture of computer, electronic and optical products': 'Computer & Optical Products'
}

oecd_df['activity'] = oecd_df['Economic activity'].map(activity_labels)
country_order = oecd_df.groupby('Country')['investment_usd'].sum().sort_values(ascending = False).index.tolist()

bar_chart = alt.Chart(oecd_df).mark_bar().encode(
    alt.Y('Country:N', sort = country_order, title = "Country"),
    alt.X('sum(investment_usd):Q', title = "Total Investment (USD)", axis = alt.Axis(format = '~s')),
    alt.Color('activity:N', title = "Economic Activity"),
    tooltip = [
        alt.Tooltip('Country:N', title = "Country"),
        alt.Tooltip('activity:N', title = "Economic Activity"),
        alt.Tooltip('sum(investment_usd):Q', title = "Total Investment (USD)", format = ',.0f')
    ]
).properties(
    width = 500,
    height = alt.Step(25),
    title = "Total Semiconductor Investment by Country and Activity (2018-2022)"
)

bar_chart
```

### Q1: The relationship between AI model performance and semiconductor investment (Dataset 8)

#### 1. Bar Chart: Top 15 Semiconductor Companies by Market Cap (2025)

```python
companies_df = pd.read_csv('data/processed/dataset8/semiconductor_companies_clean.csv')

highlight_countries = ['United States', 'South Korea', 'Taiwan', 'China']
color_scale = alt.Scale(
    domain = highlight_countries + ['Other'],
    range = ['#4C78A8', '#F58518', '#54A24B', '#E45756', '#BBBBBB']
)

top15_df = companies_df.nlargest(15, 'market_cap_usd').reset_index(drop = True)
top15_df['country_group'] = top15_df['country'].apply(
    lambda x: x if x in highlight_countries else 'Other'
)
company_order = top15_df.sort_values('market_cap_usd', ascending = False)['company'].tolist()

bar_chart = alt.Chart(top15_df).mark_bar().encode(
    alt.Y('company:N', sort = company_order, title = "Company"),
    alt.X('market_cap_usd:Q', title = "Market Cap (USD)", axis = alt.Axis(format = '~s')),
    alt.Color('country_group:N', scale = color_scale, title = "Country"),
    tooltip = [
        alt.Tooltip('company:N', title = "Company"),
        alt.Tooltip('country:N', title = "Country"),
        alt.Tooltip('market_cap_usd:Q', title = "Market Cap (USD)", format = ',.0f')
    ]
).properties(
    width = 500,
    height = 400,
    title = "Top 15 Semiconductor Companies by Market Cap (2025)"
)

bar_chart
```
