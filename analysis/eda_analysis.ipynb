### Q1: What countries have the greatest deposit? (Dataset 1)

#### 1. Bar Chart: Silicon Production by Country (2023)

```python
silicon_df = pd.read_csv('data/processed/dataset1/silicon_production_by_country.csv')
silicon_df_2023 = silicon_df[(silicon_df['material'] == 'Silicon Metal') & (silicon_df['production_2023_metric_tons'].notna())][['country', 'production_2023_metric_tons']]
country_order = silicon_df_2023.sort_values('production_2023_metric_tons', ascending = False)['country'].tolist()

bar_chart = alt.Chart(silicon_df_2023).mark_bar().encode(
    alt.Y('country:N', sort = country_order, title = "Country"),
    alt.X('production_2023_metric_tons:Q', title = "Production (metric tons)"),
    tooltip = [
        alt.Tooltip('country:N', title = "Country"),
        alt.Tooltip('production_2023_metric_tons:Q',
                    title = "Production (metric tons)", format = ',')
    ]
).properties(
    width = 400,
    height = 400,
    title = "Silicon Metal Production by Country (2023)"
)

bar_chart
```

### Q2: How much is it sold off? (Dataset 2)

#### 1. Line Chart: Price Trends by Material (2020-2024)

```python
prices_df = pd.read_csv('data/processed/dataset2/prices_clean.csv')
price_cols = {
    'Silicon_cents_per_lb': ('Silicon', 'Price (cents per lb)'),
    'Germanium_USD_per_kg': ('Germanium', 'Price (USD per kg)'),
    'Gallium_USD_per_kg': ('Gallium', 'Price (USD per kg)'),
    'Arsenic_USD_per_kg': ('Arsenic', 'Price (USD per kg)')
}

charts = []
for col, (material, y_title) in price_cols.items():
    chart = alt.Chart(prices_df).mark_line(point = True).encode(
        alt.X('Year:O', title = "Year"),
        alt.Y(f'{col}:Q', title = y_title),
        tooltip = [
            alt.Tooltip('Year:O', title = "Year"),
            alt.Tooltip(f'{col}:Q', title = y_title, format = '.2f')
        ]
    ).properties(
        width = 250,
        height = 200,
        title = material
    )
    charts.append(chart)

line_chart = alt.hconcat(
    alt.vconcat(charts[0], charts[1]),
    alt.vconcat(charts[2], charts[3])
).properties(
    title = "Price Trends by Material (2020-2024)"
)

line_chart
```

### Q3: Which countries restrict access? (Dataset 3)

#### 1. Bar Chart: Trade Interventions by Implementing and Targeted Jurisdictions

```python
interventions_df = pd.read_csv('data/processed/dataset3/interventions_clean.csv')
eu_label = 'Austria, Belgium, Bulgaria, Croatia, Cyprus, Czechia, Denmark, Estonia, Finland, France, Germany, Greece, Hungary, Ireland, Italy, Latvia, Lithuania, Luxembourg, Malta, Netherlands, Poland, Portugal, Romania, Slovakia, Slovenia, Spain, Sweden'
interventions_df['Implementing Jurisdictions'] = interventions_df['Implementing Jurisdictions'].replace(eu_label, 'European Union')

# Chart 1: implementing jurisdiction
implementing_count_df = interventions_df.groupby(
    'Implementing Jurisdictions'
).size().reset_index(name = 'count')
implementing_order = implementing_count_df.sort_values('count', ascending = False)['Implementing Jurisdictions'].tolist()

bar_chart1 = alt.Chart(implementing_count_df).mark_bar().encode(
    alt.Y('Implementing Jurisdictions:N', sort = implementing_order, title = "Implementing Country", axis = alt.Axis(labelLimit = 200)),
    alt.X('count:Q', title = "Number of Interventions"),
    tooltip = [
        alt.Tooltip('Implementing Jurisdictions:N', title = "Implementing Country"),
        alt.Tooltip('count:Q', title = "Number of Interventions")
    ]
).properties(
    width = 400,
    height = alt.Step(30),
    title = "Total Trade Interventions by Implementing Jurisdiction"
)

# Chart 2: affected jurisdiction (top 10)
affected_countries = interventions_df['Affected Jurisdictions'].dropna().str.split(', ').explode().reset_index(drop = True)
affected_count_df = affected_countries.value_counts().reset_index()
affected_count_df.columns = ['country', 'count']
affected_count_df = affected_count_df.head(10)
affected_order = affected_count_df.sort_values('count', ascending = False)['country'].tolist()

bar_chart2 = alt.Chart(affected_count_df).mark_bar().encode(
    alt.Y('country:N', sort = affected_order, title = "Affected Country"),
    alt.X('count:Q', title = "Number of Interventions"),
    tooltip = [
        alt.Tooltip('country:N', title = "Affected Country"),
        alt.Tooltip('count:Q', title = "Number of Interventions")
    ]
).properties(
    width = 400,
    height = alt.Step(30),
    title = "Top 10 Most Targeted Countries by Trade Intervention"
)

bar_chart1 | bar_chart2
```
